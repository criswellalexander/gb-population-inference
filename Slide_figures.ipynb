```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
fs = np.logspace(-4,-2,100)
Amp = 1e-21
spec_base = Amp*fs**(-2/3)*np.tanh((2e-3 - fs)/8e-4)
```

```python
sigma_log = np.linspace(0.05,0.7,100)
spec_noise = spec_base + spec_base*10**st.norm.rvs(loc=0,scale=sigma_log,size=len(spec_base))
```

```python
rng = np.random.default_rng(170815)
added_binary_fs = rng.uniform(7e-4,8e-3,size=5)
added_binary_amps = 10**rng.uniform(-18.5,-17,size=5)
```

```python
plt.figure()
plt.loglog(fs,spec_noise)
plt.scatter(added_binary_fs,added_binary_amps,marker="*")
plt.yticks([])
plt.xticks([])
# plt.xlabel("Frequency",fontsize=16)
# plt.ylabel("PSD",fontsize=16)
plt.show()
```
